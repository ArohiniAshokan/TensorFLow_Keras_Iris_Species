```python
# input x and target y split
x=df.drop(['Id','Species'],axis=1).values
x
```

```python
y=df['Species'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
# encoding target
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()  # (-1,1) here -1 indicates that the no of rows remains same
y_train_encoded=encoder.fit_transform(y_train.reshape(-1,1)).toarray()   # toarray()-to convert sparse matrix to dense array
y_test_encoded=encoder.transform(y_test.reshape(-1,1)).toarray()

# sparse matrix is a matrix where only a small percentage of its elements are non-zero.
# sparse matrices typically store only the non-zero elements and their indices.

# In contrast to sparse matrices, dense arrays store all elements of a matrix, regardless of whether
# they are zero or non-zero. This means that even zero elements consume memory in a dense array.

# This conversion is not strictly necessary for training a neural network,
# but it's often done for convenience, as many libraries (including TensorFlow/Keras) expect dense arrays as input.
```

```python
# The shape (105, 3) of y_train_encoded indicates that there are 105 samples (rows) and 3 columns.
# Each column corresponds to one of the three classes in your dataset.

# The reason for this shape is that OneHotEncoder automatically determines the number of columns based
# on the unique categories present in the target labels (y_train). Since you have three classes in your dataset,
# OneHotEncoder creates three columns in the one-hot encoded representation, one for each class.

# When you reshape y_train into a 2D array with a single column using y_train.reshape(-1, 1), you're preparing
# the target labels for one-hot encoding by converting them into a format where each row contains the label of a single sample.

# However, during the one-hot encoding process, OneHotEncoder examines the unique categories in y_train and creates a
# binary column for each unique category. In your case, there are three unique categories (classes),
# so OneHotEncoder generates three columns in the resulting one-hot encoded representation.

# Therefore, even though you reshaped y_train to have a single column, the one-hot encoding process
# itself determines the number of columns based on the unique categories present in the data.

# In summary, the shape (105, 3) indicates that there are 105 samples and 3 columns in the one-hot encoded representation,
# with each column representing one of the three classes in your dataset.
```

```python
print(x_train.shape)
print(y_train_encoded.shape)
print(x_test.shape)
print(y_test_encoded.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

For simpler classification/regression tasks or smaller datasets, you might start with a smaller number of neurons in the hidden layers, such as 8 to 128 neurons per layer.
For more complex classification/regression tasks or larger datasets, you might use a larger number of neurons, such as 128 to 512 neurons per layer.

For tabular data, a dataset with a shape like (100, 10) or (500, 20) could be considered relatively small.
For image data, small datasets might have images with resolutions like 64x64 pixels, resulting in a shape like (num_samples, 64, 64, num_channels).

```python
# defining architecture of our NN
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(4, )))   #x_train.shape[1] is 4 ie inputs no of columns
# this layer acts as both the input layer and the first hidden layer.
# When you specify input_shape in the first layer of a Sequential model, Keras automatically creates the input layer for you.

model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
# compile the model

# "compiling the model" refers to configuring the model for training
# by specifying additional settings required for the training process. This includes specifying the optimizer, loss function,
# and metrics that will be used to evaluate the model's performance during training.

model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy']) # Adam - Adaptive Moment Estimation
model.summary()

# Trainable parameters are the weights and biases in the neural network that are updated during the training process.
# non-trainable parameters are fixed and do not change during training
```

"verbose" in the compile function typically refers to a parameter that controls the amount of information printed during the training process.

verbose=0: Silent mode. No output during training.

verbose=1: Default mode. Progress bars displaying training progress and metrics.

verbose=2: Verbose mode. One line per epoch showing training progress and metrics.

```python
model.fit(x_train, y_train_encoded, epochs=100, verbose=2)
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(x_test, y_test_encoded)
print(f'Test Loss: {loss}')
print(f'Test Accuracy: {accuracy}')
```

```python
y_pred=model.predict(x_test)    # these are all probabilities due to softmax activation fn
y_pred
```

```python
y_pred.shape
```

```python
#  np.argmax() is being used on an array of probabilities (y_pred) rather than on a 2D array representing
# multiple samples and their predictions.

# The np.argmax() function is intended to find the index of the maximum value along a specified axis of an array.
#  When applied to an array of probabilities, it returns the index of the class with the highest probability for each sample.

# shape of y_pred is like (batch_size, num_classes), where batch_size represents
# the number of data points in the batch, and num_classes represents the number of classes.
# In this case, the last axis (axis -1) of y_pred corresponds to the different classes.
```

```python
import numpy as np
y_pred=np.argmax(y_pred,axis=-1)
y_pred
```

```python
encoder.categories_[0]
```

```python
# encoder.categories_[0] contains the original categorical values and y_pred contains
# the predicted numerical labels, encoder.categories_[0][y_pred] will transform the numerical predictions
# back into their original categorical values.
```

```python
y_pred_classes = encoder.categories_[0][y_pred]
y_pred_classes
```

```python
y_test
```

```python
df_pred=pd.DataFrame({'y_actual':y_test,'y_predicted':y_pred_classes})
df_pred
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred_classes)
cm
```

```python
cmd=ConfusionMatrixDisplay(cm,display_labels=df['Species'].unique())
cmd.plot()
```

```python
report=classification_report(y_test,y_pred_classes)
print(report)
```
